```python
scoreabovezero=df['Score']>'0-0'
```

```python
df[scoreabovezero].shape or df.shape
```

```python
scorebyBscore=df['home_team']=='Brazil'
```

```python
brazildf=df[scoreabovezero & scorebyBscore].sort_values(by='home_score', ascending=True)
```

```python
brazildf.info()
```

```python
%matplotlib inline
```

```python
from matplotlib import pyplot as plt
```

```python
brazildf.plot(x='home_score',y='away_score',kind='scatter')
plt.xlabel('Home Score')
plt.ylabel('Away Score')
plt.legend('Take Away Score')
plt.title('Brazil Matches')
plt.show()
```

```python
brazildf.plot(x='home_score',y='away_score',kind='bar')
plt.xlabel('Home Score')
plt.ylabel('Away Score')
plt.legend('Take Away Score')
plt.title('Brazil Matches')
plt.show()
```

```python
brazildf.plot(x='home_score',y='away_score',kind='pie')
plt.xlabel('Home Score')
plt.ylabel('Away Score')
plt.legend('Take Away Score')
plt.title('Brazil Matches')
plt.show()
```
